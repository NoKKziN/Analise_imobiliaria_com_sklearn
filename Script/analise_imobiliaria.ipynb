```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df = pd.read_csv("sao-paulo-properties-april-2019.csv")
```

```python
df.head()
```

```python
df["Negotiation Type"].value_counts()
```

```python
df_rent = df[df["Negotiation Type"] == "rent"]
```

```python
df_cleaned = df.drop(["New", "Negotiation Type", "Property Type"], axis=1)

df_cleaned
```

```python
one_hot = pd.get_dummies(df_cleaned["District"])

df = df_cleaned.drop("District", axis=1)
df = df.join(one_hot)
```

```python
df
```

```python
y = df["Price"]
X = df.loc[:, df.columns != "Price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
X_train
```

```python
y_train
```

```python
# Usando um modelo de regressão linear para prever os valores possiveis
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
alguns_dados = X_test.iloc[:5]
algumas_label = y_test.iloc[:5]

print("Predições:", lin_reg.predict(alguns_dados))
print("Labels:", algumas_label.values)
```
